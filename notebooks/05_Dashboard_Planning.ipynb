# Key Insights Highlighted in the Dashboard

1. **Content Explosion Post-2015:** Netflix's catalog expanded rapidly after 2015, reflecting the platform's aggressive content acquisition and production strategy during its global expansion phase. Titles peaked at 1,121 in 2018.

2. **Movies Dominate, TV Shows Rising:** Although Movies comprise ~69% of the catalog (5,377 titles), the share of TV Shows has increased steadily, with TV Shows surpassing Movies in new additions by 2020 (457 vs 411).

3. **Geographic Concentration:** The top 5 countries (United States: 3,296, India: 990, United Kingdom: 722, Canada: 412, France: 349) contribute the majority of content. The US alone accounts for ~42% of all titles.

4. **Mature Content Focus:** Over 60% of titles are rated for mature audiences (TV-MA: 2,863 + TV-14: 1,931), indicating the platform primarily targets adult viewers.

5. **International Content is King:** "International Movies" (2,437) is the largest genre, followed by "Dramas" (2,106) and "Comedies" (1,471) — reflecting Netflix's global content strategy.

6. **Country × Type Association:** A statistically significant but weak association exists between country and content type (χ² = 214.67, p < 0.001, Cramér's V = 0.193). Country influences content mix but doesn't strongly determine it — e.g., India skews heavily toward Movies (915 vs 75 TV Shows), while Japan favors TV Shows (184 vs 103 Movies).

7. **Bollywood Dominance in Casting:** The most frequently appearing actors are predominantly from Indian cinema — Anupam Kher (42), Shah Rukh Khan (35), Om Puri (30) — reflecting India's outsized contribution to movie content.

8. **Data Quality Fix:** The original country field counted co-produced titles (e.g., "United Kingdom, United States") as their own category. After exploding and cleaning, the true unique country count dropped from 681 to 121, revealing the United Kingdom's actual contribution was nearly double what raw data showed (397 → 722).
